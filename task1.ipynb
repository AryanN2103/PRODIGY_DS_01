```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/diabetes.csv")
```

```python
df.head
```

```python
df['Glucose'].plot.hist(bins=5, color='blue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of the Dataset')

plt.show()
```
